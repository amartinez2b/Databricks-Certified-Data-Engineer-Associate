
# Fundamentos de la gobernanza de datos y su manejo en Databricks

La gobernanza de datos consiste en establecer políticas, procesos y controles para asegurar que los datos sean precisos, seguros, accesibles y auditables durante todo su ciclo de vida. Sus fundamentos incluyen:

- **Unificación de datos**: Centraliza y organiza todos los activos de datos y de IA, evitando duplicidad y dispersión.
- **Control de acceso**: Permite que solo los usuarios autorizados accedan a los datos, mediante modelos jerárquicos de privilegios.
- **Descubrimiento y calidad**: Facilita encontrar los datos necesarios y garantiza que sean completos, consistentes y seguros.
- **Colaboración y compartición**: Permite compartir datos de forma segura dentro y fuera de la organización.
- **Auditoría**: Registra quién accede y qué acciones realiza sobre los datos.

En Databricks, la gobernanza de datos se gestiona principalmente a través de **Unity Catalog**, que proporciona:

- Organización centralizada de datos en catálogos, esquemas y tablas.
- Control granular de permisos y acceso, desde el nivel de cuenta hasta filas y columnas.
- Herramientas para trazabilidad y auditoría mediante logs y tablas de sistema.
- Integración con flujos de trabajo de datos y machine learning, asegurando cumplimiento y seguridad en la nube.

# ¿Qué es Unity Catalog y para qué sirve?

Unity Catalog es una solución unificada de gobernanza para activos de datos y de IA en Databricks.  
Sirve para administrar, asegurar y auditar el acceso a datos, modelos y otros objetos en la plataforma, permitiendo:

- Organización centralizada de datos en catálogos, esquemas y tablas, facilitando la administración y el descubrimiento.
- Control de permisos y acceso a los datos desde un solo lugar, mejorando la seguridad y el cumplimiento normativo.
- Trazabilidad y auditoría de todas las acciones realizadas sobre los datos.
- Integración con flujos de trabajo de datos y machine learning en Databricks.

Unity Catalog es esencial para empresas que requieren una gestión robusta y segura de sus activos de datos en la nube.


# Estructura de objetos en Unity Catalog

Unity Catalog organiza los activos de datos en una jerarquía clara y centralizada:

- **Catálogo**: Es el nivel superior, agrupa esquemas y define límites de gobernanza.
- **Esquema (Schema)**: Dentro de cada catálogo, los esquemas organizan tablas, vistas y otros objetos relacionados.
- **Tabla/Vista**: Son los objetos de datos donde se almacenan y consultan los datos. Las tablas pueden ser gestionadas o externas.
- **Funciones y modelos**: También pueden almacenarse y gestionarse dentro de esquemas.

Esta estructura permite aplicar controles de acceso y auditoría en cada nivel, facilitando la administración, seguridad y descubrimiento de los datos en Databricks.
